#### 1. Mengautentikasi Twitter

```python
import tweepy as tw
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
consumer_key = '*************************'
consumer_key = '*************************'
consumer_secret = '**************************************************'
access_token = '**************************************************'
access_token_secret = '*********************************************'
```

```python
auth = tw.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tw.API(auth, wait_on_rate_limit = True)
```

#### 2. Ambil Data Twitter

```python
hashtag = '#disney'
query = tw.Cursor(api.search_tweets, q=hashtag).items(2000)
tweets = [{'Tweets':tweet.text, 'Timestamp':tweet.created_at} for tweet in query]
```

```python
df = pd.DataFrame.from_dict(tweets)
df.head()
```

```python
chapek_refs = ['BobChapek','Bob Chapek','Chapek','Chapek\s']
iger_refs = ['BobIger','Bob Iger','Iger','Iger\s']
```

```python
def identify_subject(tweet, refs):
    flag = 0
    for ref in refs:
        if tweet.find(ref) !=-1:
            flag = 1
    return flag

df['Chapek'] = df['Tweets'].apply(lambda x: identify_subject(x, chapek_refs))
df['Iger'] = df['Tweets'].apply(lambda x: identify_subject(x, iger_refs))
```

```python
df.head()
```

#### 3. Prapemrosesan

```python
pip install -U textblob
```

```python
#import stopwords
import nltk
from nltk.corpus import stopwords

# import textblob
from textblob import Word, TextBlob
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
stop_words = stopwords.words('english')
custom_stopwords = ['RT', '#onelove']
```

```python
def preprocess_tweets(tweet, custom_stopwords):
    preprocessed_tweet = tweet
    preprocessed_tweet.replace('[^\w\s]', '')
    preprocessed_tweet = " ".join(word for word in preprocessed_tweet.split() if word not in stop_words)
    preprocessed_tweet = " ".join(word for word in preprocessed_tweet.split() if word not in custom_stopwords)
    preprocessed_tweet = " ".join(Word(word).lemmatize() for word in preprocessed_tweet.split())
    return(preprocessed_tweet)

df['Preprocessed Tweet'] = df['Tweets'].apply(lambda x: preprocess_tweets(x, custom_stopwords))
df.head()
```

```python
df[df['Chapek']==1]
```

```python
df[df['Iger']==1]
```

#### 4. Menghitung Sentiment

```python
df['polarity'] = df['Preprocessed Tweet'].apply(lambda x: TextBlob(x).sentiment[0])
df['subjectivity'] = df['Preprocessed Tweet'].apply(lambda x: TextBlob(x).sentiment[1])
```

```python
df.tail()
```

```python
display(df[df['Chapek']==1][['Chapek','polarity','subjectivity']].groupby('Chapek').agg([np.mean, np.max, np.min, np.median]))
df[df['Iger']==1][['Iger','polarity','subjectivity']].groupby('Iger').agg([np.mean, np.max, np.min, np.median])
```

#### 5. Visualisasi

```python
chapek = df[df['Chapek']==1][['Timestamp','polarity']]
chapek = chapek.sort_values(by='Timestamp', ascending=True)
chapek['MA Polarity'] = chapek.polarity.rolling(10, min_periods=3).mean()

iger = df[df['Iger']==1][['Timestamp','polarity']]
iger = iger.sort_values(by='Timestamp', ascending=True)
iger['MA Polarity'] = iger.polarity.rolling(10, min_periods=3).mean()
```

```python
chapek.head()
```

```python
iger.head()
```

```python
fig, axes = plt.subplots(2,1, figsize=(13,10))

axes[0].plot(chapek['Timestamp'], chapek['MA Polarity'])
axes[0].set_title("\n".join(['Chapek 10 Tweet Moving Average Polarity']))

axes[1].plot(iger['Timestamp'], iger['MA Polarity'], color='red')
axes[1].set_title("\n".join(['Iger 10 Tweet Moving Average Polarity']))

fig.suptitle('\n'.join(['Pergantian CEO Disney, Bob Chapek menjadi Bob Iger']), y=0.98)

plt.show()
```
